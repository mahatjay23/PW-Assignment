Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

What is Matplotlib?

Matplotlib is a widely-used open-source Python library for creating static, interactive, and animated visualizations. It provides a flexible framework for generating a variety of plots and charts, making it a fundamental tool for data visualization in Python.

Why is it Used?

Matplotlib is used because:

Data Visualization: It helps in visualizing data to uncover patterns, trends, and insights.
Flexibility: Offers extensive customization options for creating tailored plots.
Integration: Works seamlessly with other Python libraries like NumPy and pandas.
Publication-Quality Graphics: Capable of producing high-quality figures suitable for publications.
Interactive Features: Supports interactive plots for exploratory data analysis.

Five Plots Using Pyplot Module of Matplotlib:

1. Line Plot: Displays data points connected by lines, ideal for showing trends over time.
2. Scatter Plot: Plots individual data points on a two-dimensional plane, useful for identifying relationships between variables.
3. Bar Chart: Represents categorical data with rectangular bars, effective for comparing quantities across categories.
4. Histogram: Shows the distribution of numerical data by grouping data points into bins and displaying their frequencies.
5. Pie Chart: Illustrates proportions of a whole as slices of a circular pie, useful for showing percentage distributions.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot

What is a Scatter Plot?

A scatter plot is a type of data visualization that displays points on a two-dimensional plane, where each point represents the values of two variables. The horizontal axis (x-axis) represents one variable, and the vertical axis (y-axis) represents the other. Scatter plots are useful for showing the relationship or correlation between two variables. Patterns, trends, or clusters in the data can often be observed in scatter plots.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y,s=100,color="red")
plt.xlabel("x-->axis")
plt.ylabel("y-->axis")
plt.title("Scatter Plot of Generated Data")
plt.show()
```

Why is the subplot() Function Used?

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to divide a figure into a grid of subplots, where each subplot can contain a different plot. This is particularly useful for comparing multiple plots side by side or for organizing several related visualizations in a compact space.

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.xlabel("x-->axis")
plt.ylabel("y-->axis")
plt.plot(x,y)

plt.subplot(2,2,2)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.xlabel("x-->axis")
plt.ylabel("y-->axis")
plt.plot(x,y)

plt.subplot(2,2,3)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.xlabel("x-->axis")
plt.ylabel("y-->axis")
plt.plot(x,y)

plt.subplot(2,2,4)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.xlabel("x-->axis")
plt.ylabel("y-->axis")
plt.plot(x,y)

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

What is a Bar Plot?

A bar plot (or bar chart) is a graphical representation of categorical data using rectangular bars. The length or height of each bar is proportional to the value it represents. Bar plots are commonly used to compare the quantities of different categories or groups.

Why is a Bar Plot Used?

Bar plots are used because:

They make it easy to compare the values of different categories.
1. Clarity: The visual length of the bars makes differences between categories clear.
2. Versatility: Bar plots can be oriented vertically or horizontally, making them adaptable to various presentation needs.
3. Categorical Data: Ideal for representing and comparing discrete categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit,color=["red","yellow","pink","blue"],edgecolor="black")
plt.xlabel("Company_Name--->")
plt.ylabel("Profit--->")
plt.grid()
plt.title("Horizontal Bar Plot")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

What is a Box Plot?

A box plot, also known as a box-and-whisker plot, is a graphical representation that summarizes the distribution of a dataset. It displays the median, quartiles, and potential outliers, offering a visual insight into the spread and symmetry of the data.

Why is it Used?

Box plots are used to:

1. Visualize the distribution of a dataset.
2. Compare distributions across different datasets.
3. Identify outliers and understand data variability.
4. Summarize key statistical measures like the median and quartiles.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title("Box Plot of Two Distributions")
plt.ylabel("Values")
plt.show()
```
